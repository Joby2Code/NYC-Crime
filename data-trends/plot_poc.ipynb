```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

#plotting on the canvas
x=[1,2,3]
y=[4,5,3]

plt.plot(x,y,label='coordinates',linewidth=2,color='b')
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('POC')

plt.legend()
plt.grid(True,color='k')
plt.show()

```

```python
#Plot different graphs


from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

#plotting on the canvas
x=[1,2,3]
y=[4,5,10]

plt.bar(x,y,label='coordinates',color='b')
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('POC')

plt.legend()
plt.grid(True,color='g')
plt.show()
```

```python
#Handling multiple plots
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1=np.arange(0.0,5,0.5)
t2=np.arange(0.0,5,1)

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.plot(t1,f(t1))
plt.show()
```

```python
#Handling multiple plots
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1=np.arange(0.0,5,0.5)
t2=np.arange(0.0,5,1)

t3=np.arange(0.0,5,0.1)
t4=np.arange(0.0,7,0.6)

plt.subplot(311)
plt.plot(t2,f(t2))
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(312)
plt.plot(t3,np.cos(2*np.pi*t3))

plt.subplot(313)
plt.plot(t4,f(t4))



plt.show()
```
