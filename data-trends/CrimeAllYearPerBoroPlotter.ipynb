```python
import re
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np

borough='STATEN ISLAND'

def plotgraph(data):
    lists = sorted(data.items()) # sorted by key, return a list of tuples
    year, frequency = zip(*lists) # unpack a list of pairs into two tuples
    
    #Plot data
    fig = plt.figure(figsize = (20,3))
    plt.title('Crime Occurences Per Year')
    plt.ylabel('Crime commited/reported')
    plt.xlabel('Year')
    plt.plot(year, frequency, marker='o',label=str(borough),color='b')
    plt.legend()
    fig.savefig("crime-plots-borough/crime_all_year_per"+str(borough)+".png")  
    plt.show()

with open("crime-data-by-borough/crime-bourough-per-year.out") as ins:
    year_boro_dict={}
    
    for line in ins:
        tuple=line.replace("'","").replace('(',"").replace(')',"").split(',')
        boro_freq_dict={}
        yr=int(tuple[0])
        boro=tuple[1].strip()
        freq=int(tuple[2])
      
        if boro in year_boro_dict:
            boro_freq_dict = year_boro_dict.get(boro)  
            boro_freq_dict[yr]=freq
            year_boro_dict[boro]=boro_freq_dict
        else:     
            boro_freq_dict[yr]=freq
            year_boro_dict[boro]=boro_freq_dict     
    try:
        data = year_boro_dict.get(borough)
        plotgraph(year_boro_dict.get(borough))
    except Exception as e:
        print('Enter the correct Borough')
        
    year_boro_dict={}
```
